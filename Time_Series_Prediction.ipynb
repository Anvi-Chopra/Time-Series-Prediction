```python
import pandas as pd 
import numpy as np          
import matplotlib.pyplot as plt   
```

```python
df=pd.read_csv("dataset road.csv")
```

```python
df2=df.iloc[9::10,:]
df2.head()
```

```python
df2.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df2.shape
```

```python
df.drop(df.index[9::10], axis=0, inplace=True)
```

```python
valid=df2.reset_index()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
import math
from sklearn.metrics import mean_squared_error
```

```python
predictions1=list()
y=0
for i in range(0,100):
  z=y+9
  train=df.iloc[y:z]
  #print(train.head(10))
  ARMAmodel = ARIMA(train['Para-9'], order = (1, 1, 0))
  ARMAmodel = ARMAmodel.fit()
  output = ARMAmodel.forecast(1)
  predictions1.append(output)
  y=z
```

```python
rms = math.sqrt(mean_squared_error(valid['Para-9'], predictions1)) 
print(rms)
```

```python
plt.plot(valid['Para-9'], label='Valid') 
plt.plot(predictions1, label='Predictions') 
plt.legend(loc='best') 
plt.show()
```
